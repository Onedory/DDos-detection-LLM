```python
from tqdm import tqdm  # 진행 바 라이브러리
# 8개의 텍스트 파일 불러오기
file_paths = [f'/content/drive/MyDrive/LLM/text_data{i}.txt' for i in range(1, 9)]

# 파일에서 데이터를 읽어와서 리스트에 저장
text_data = []
labels = []

# tqdm을 사용하여 진행률 확인
for file_path in tqdm(file_paths, desc="Loading text files", unit="file"):
    with open(file_path, 'r') as f:
        for line in f:
            if 'Label: BENIGN' in line:  # 수정된 부분
                labels.append(0)  # 정상 트래픽
            else:
                labels.append(1)  # 비정상 트래픽 (DDoS)
            text_data.append(line.strip().split(' -->')[0])  # 레이블을 제거한 문장만 남김

# 데이터 확인
print(f"Total data loaded: {len(text_data)}")
print(f"BENIGN count: {labels.count(0)}, DDoS count: {labels.count(1)}")
```

```python
import pickle
from sklearn.model_selection import train_test_split

# 1. 텍스트 데이터와 레이블을 학습용과 검증용으로 분할
train_texts, val_texts, train_labels, val_labels = train_test_split(text_data, labels, test_size=0.2, random_state=42)

# 2. 분할된 데이터를 Google Drive에 저장하기
with open('/content/drive/MyDrive/LLM/data/train_texts.pkl', 'wb') as f:
    pickle.dump(train_texts, f)

with open('/content/drive/MyDrive/LLM/data/val_texts.pkl', 'wb') as f:
    pickle.dump(val_texts, f)

with open('/content/drive/MyDrive/LLM/data/train_labels.pkl', 'wb') as f:
    pickle.dump(train_labels, f)

with open('/content/drive/MyDrive/LLM/data/val_labels.pkl', 'wb') as f:
    pickle.dump(val_labels, f)

print("train_texts, val_texts, train_labels, val_labels saved successfully!")
```
