```python
import zipfile
import os

# 업로드된 파일의 이름이 'archive.zip'이라고 가정
zip_file_path = "/content/drive/MyDrive/LLM/archive.zip"
unzip_dir = "/content/drive/MyDrive/LLM"

# 압축 해제
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_dir)

print(f"압축이 풀린 경로: {unzip_dir}")
```

```python
# 압축 해제된 파일 목록 출력
for root, dirs, files in os.walk(unzip_dir):
    for file in files:
        print(os.path.join(root, file))
```
