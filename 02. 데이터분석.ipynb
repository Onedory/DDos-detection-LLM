```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/drive/MyDrive/LLM/Friday-WorkingHours-Morning.pcap_ISCX.csv')

df.head()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df[' Label']= encoder.fit_transform(df[' Label'])
```

```python
df.head()
```

```python
# 0과 1 각각의 개수를 확인합니다
print(df[' Label'].value_counts())
```

```python
df = df.fillna(0)  # Replace NaN with 0
df = df.replace([np.inf, -np.inf], 0) #양/음의 무한대를 0으로 변경, 수학적 오류나 나눗셈 연산에서 무한대값이 발생할 수 있음
```

```python
df.isnull().sum()
```

```python
df=df.astype(int) #데이터프레임의 모든 값을 정수형으로 변환
```

```python
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif

X = df.drop(' Label',axis=1) #Label 열(정답 레이블)을 제외한 모든 열을 입력 데이터로 사용
y = df[' Label'] #Lable 열만 따로 저장하여 정답 레이블로 사용
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #각 특성을 평균이 0, 표준편차가 1이 되도록 표준화한다.
#이유: 머신러닝 알고리즘(예: SVM, KNN)은 특성의 스케일에 민감하므로, 모든 특성을 동일한 스케일로 맞추는 것이 중요합니다.
X_scaled = scaler.fit_transform(X) #fit(): 입력 데이터의 평균과 표준편차를 게산, transform(): 계산한 값으로 데이터를 변환
```

```python
from sklearn.impute import SimpleImputer #simpleimputer는 결측값(nan)을 특정값으로 대체
# Impute missing values (replace NaNs with the mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X) #fit: 각 열의 평균값 계산, transform: 계산한 평균값으로 결측값 채우기

# Determine the number of columns (features) in your DataFrame
num_columns = df.shape[1] #데이터프레임의 열 개수(특성 수)를 반환

# Set an appropriate value for k (less than or equal to the number of columns)
k = min(10, num_columns)  # Adjust this as needed, k는 상위 k개의 중요한 특성을 선택하는데 사용
#min(10, num_columns)는 특성개수가 10개 이상인 경우 상위 10개만 선택하고 그렇지않으면 전체열 사용

# Initialize SelectKBest with the scoring function
k_best = SelectKBest(score_func=f_classif, k=k) #selectKBest는 상위 k개의 중요한 특성만 선택하는 특성 선택 기법
#scroe_func = f_classif : 분류 문제에 적합한 ANOVA F-통계량을 사용해 각 특성의 중요도 평가

# Fit and transform the imputed data to select the top 10 features
X_new = k_best.fit_transform(X_imputed, y)
#fit_transform(): 각 특성의 중요도를 계산한 후, 가장 중요한 상위 K개의 특성만 선택한다.
```

```python
# Get the boolean mask of selected features
selected_features_mask = k_best.get_support()
#get_support()는 selectKBest가 선택한 상위 K개의 특성에 대한 마스크(불리언 값)을 반환
#이 마스크는 True/False값으로 구성되며, 선택된 특성은 True로 나타낸다
```

```python
elected_feature_names = X.columns[selected_features_mask] #get_support()에서 반환된 불리언 마스크를 사용해 선택된 특성의 이름만 추출
```

```python
elected_feature_names
#출력된 결과는 SelectKBest가 선택한 상위10개의 중요한 특성 이름이다.
#이들 특성은 ANOVA F-통계량(f_classif)을 기준으로 가장 레이블과 높은 상관관계를 가진 상위 10개의 특성
```

```python
new_columns=[' Destination Port', ' Bwd Packet Length Min',
       ' Bwd Packet Length Mean', ' Bwd Packets/s', ' Min Packet Length',
       ' PSH Flag Count', ' URG Flag Count', ' Avg Fwd Segment Size',
       ' Avg Bwd Segment Size', ' min_seg_size_forward']
```

```python
df_new=X[new_columns]
#원래의 데이터프레임 x에서, 선택된 10개의 특성만 포함한 새로운 데이터프레임을 생성한다.
#즉 df_new는 상위 10개의 특성만 포함한 새로운 데이터프레임이다.
```

```python
df_new
```

```python
df_new['label']=df[' Label']
#df_new에 선택된 10개의 특성(원래있던거)과 레이블(정답 값)을 포함하도록 수정
```

```python
#특성(x1)과 레이블(y1)분리
X1=df_new.iloc[:,:-1].values #모들행과 마지막 열을 제외(:-1)한 특성 데이터만 선택
y1=df_new.iloc[:,-1].values #모든행과 마지막 열(-1)인 레이블 데이터를 선택
#values: Pandas 데이터프레임을 numpy 배열로 변환한다. 이렇게 변환된 배열은 머신러닝 모델에 적합한 형태
```

```python
X1 #선택된 10개의 특성만을 포함한 배열
```

```python
y1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.3, random_state=42)
#random_state는 데이터를 항상 같은 방식으로 나누기 위해 사용된다
#train_test_split는 전체데이터를 훈련세트(70%), 테스트세트(30%)로 분할한다
#x_train, y_train: 모델을 학습시키기 위한 훈련데이터, x_test, y_test:모델의 성능을 평가할 테스트 데이터
```

```python
#인공신경망(ANN) 모델 구축
import tensorflow as tf
ann = tf.keras.models.Sequential() #순차적으로 레이어를 쌓는 방식의 신경망을 만든다
ann.add(tf.keras.layers.Dense(units=10,activation='sigmoid'))
#Dense 레이어는 각 노드가 전 노드와 완전히 연결된 층
#units=10 : 각 층에 10개의 뉴런을 포함.
#activation='sigmoid' : 시그모이드 활성화 함수를 사용해 출력값을 0과 1사이로 변환한다.
ann.add(tf.keras.layers.Dense(units=10,activation='sigmoid'))
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))#마지막레이어 : units=1로 설정하여 이진분류를 위해 1개의 출력값을 반환

ann.compile(optimizer ='adam' , loss='binary_crossentropy' , metrics= ['accuracy'])# 정확도(accuracy)를 성능평가 지표로 사용
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10)
#지정된 val_loss(검증 손실)가 10번 이상 감소하지 않으면 학습을 중단한다.
```

```python
#training the ann on training set
ann.fit(X_train,y_train,batch_size=32,epochs=10,callbacks=[early_stopping])
```
