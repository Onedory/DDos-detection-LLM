```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/drive/MyDrive/LLM/Friday-WorkingHours-Morning.pcap_ISCX.csv')

df.head()
```

```python
# 텍스트 데이터로 변환
text_data = []
for index, row in df.iterrows():
    # 한 행의 모든 열을 '열 이름: 값' 형태로 변환하여 텍스트로 구성
    text_entry = ', '.join([f"{col}: {row[col]}" for col in df.columns])
    text_data.append(text_entry)

# 변환된 텍스트 출력 (예시)
print(text_data[0])  # 첫 번째 행을 텍스트로 변환한 결과
```

```python
# Google Drive 경로에 텍스트 파일로 저장
file_path = '/content/drive/MyDrive/LLM/text_data1.txt'

# 파일에 쓰기
with open(file_path, 'w') as f:
    for line in text_data:
        f.write(line + '\n')  # 각 행을 한 줄로 저장
```

```python
# 저장된 텍스트 파일 불러오기
file_path = '/content/drive/MyDrive/LLM/text_data1.txt'

with open(file_path, 'r') as f:
    lines = f.readlines()  # 모든 라인을 읽어옵니다.

# 첫 5개의 라인 출력하여 변환 내용 확인
for i in range(5):
    print(lines[i])
```

```python
import os

# 파일 크기 확인
file_size = os.path.getsize(file_path)
print(f"File size: {file_size / (1024 * 1024):.2f} MB")
```
